This code produces random data that adheres to a linear pattern with added noise. It then generates a DataFrame to hold the data and shows the initial rows of the DataFrame. This code snippet serves as a simple demonstration of generating and displaying data using Python libraries.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
np.random.seed(42)  
data_size = 50
x = np.linspace(0, 10, data_size)
y = 2 * x + 1 + np.random.normal(0, 1, data_size) 
```

```python
df = pd.DataFrame({'X': x, 'Y': y})
```

```python
df.head()
```

This code produces a scatter plot using the data points generated from the DataFrame df and adds the actual relationship line. The plot is then personalized with a title, axis labels, and a legend. This enables the visualization of the generated data and comparison with the true relationship line.

```python
plt.figure(figsize=(8, 6))
plt.scatter(df['X'], df['Y'], label='Noise')
plt.plot(x, 2*x + 1, color='green', linestyle='-', label='True Relationship')
plt.title('Generated Data and True Relationship')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```

### Question 5 a. How many columns and rows does the data have? 

```python
import pandas as pd

titanic = pd.read_csv('https://raw.githubusercontent.com/mwaskom/seaborn-data/master/titanic.csv')

num_rows, num_columns = titanic.shape
print(f'The Titanic dataset has {num_rows} rows and {num_columns} columns.')
```

### Question 5 b. Get a sense of your data and find the min, max, and count/mean depending on the
data type. 

```python
numerical_summary = titanic.describe()

categorical_summary = titanic.describe(include='object')

print('Summary Statistics for Numerical Columns:')
print(numerical_summary)

print('\nSummary Statistics for Categorical Columns:')
print(categorical_summary)
```

### Question 5 c. Give an overview (code and an explanation) of all missing values in the data. 

```python
missing_values = titanic.isnull().sum()

columns_with_missing_values = missing_values[missing_values > 0]

print('Overview of Missing Values:')
print(missing_values)

print('\nColumns with Missing Values:')
print(columns_with_missing_values)
```

### Question 5 d. Delete the rows where you do not have information about the age of the person.
Then group the passengers  in a  10  year  age range (for example, you can do
something like 0 – 10, 11 – 20, 21 – 30, etc). 

```python
titanic_cleaned = titanic.dropna(subset=['age']).copy()
age_ranges = pd.cut(titanic_cleaned['age'], bins=range(0, 101, 10), right=False)
titanic_cleaned.loc[:, 'Age Range'] = age_ranges

print(titanic_cleaned.head())
```

### Question 5 e. For   each   age   category   created   in   d),   find   out   how   many   passengers   are
female/male, and how many travelled in each class.

```python
age_group_stats = titanic_cleaned.groupby(['Age Range', 'sex', 'class'], observed=True).size().unstack()

print('Passenger Statistics by Age Range, Gender, and Class:')
print(age_group_stats)
```
